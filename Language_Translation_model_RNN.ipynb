```python
# function to create a tokenizer
from keras.preprocessing.text import Tokenizer
def create_tokenizer(lines):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(lines)
    return tokenizer
```

```python
# maximum sentence length
def max_len(lines):
    return max(len(line.split()) for line in lines)
```

```python
# encode and pad sequences
from keras.preprocessing.sequence import pad_sequences
def encode_sequences(tokenizer, length, lines):
    # integer encode sequences
    seq = tokenizer.texts_to_sequences(lines)
    # pad sequences with 0 values
    seq = pad_sequences(seq, maxlen=length, padding='post')
    return seq
```

```python
# one hot encode target sequence
from keras.utils import to_categorical
import numpy as np

def encode_output(sequences, vocab_size):
    ylist = list()
    for sequence in sequences:
        encoded = to_categorical(sequence, num_classes=vocab_size)
        ylist.append(encoded)
    y = np.array(ylist)
    y = y.reshape(sequences.shape[0], sequences.shape[1], vocab_size)
    return y
```

```python
source_str, target_str = "Italian", "English"

# Index in the numpy array of the source and target
idx_src, idx_tar = 1, 0
```

```python
# Prepare target tokenizer
tar_tokenizer = create_tokenizer(ita_eng[:, idx_tar])
tar_vocab_size = len(tar_tokenizer.word_index) + 1
tar_length = max_len(ita_eng[:, idx_tar])
print(f'\nTarget ({target_str}) Vocabulary Size: {tar_vocab_size}')
print(f'Target ({target_str}) Max Length: {tar_length}')

# Prepare source tokenizer
src_tokenizer = create_tokenizer(ita_eng[:, idx_src])
src_vocab_size = len(src_tokenizer.word_index) + 1
src_length = max_len(ita_eng[:, idx_src])
print(f'\nSource ({source_str}) Vocabulary Size: {src_vocab_size}')
print(f'Source ({source_str}) Max Length: {src_length}\n')
```

```python
from sklearn.model_selection import train_test_split
# train, test = train_test_split(ita_eng_arr, test_size=0.2, random_state = 12)
train, test = train_test_split(ita_eng, test_size=0.2, random_state = 12)
```

```python
# Prepare training data
trainX = encode_sequences(src_tokenizer, src_length, train[:, idx_src])
trainY = encode_sequences(tar_tokenizer, tar_length, train[:, idx_tar])
trainY = encode_output(trainY, tar_vocab_size)
```

# Why Seq2Seq model?

In a normal Recurrent Neural Network (RNN), it takes in an input sequence and outputs a sequence. The vanishing gradient problem reduces the effectiveness of RNNs as they often will not be able to fully take into account the long sequences. To overcome this Long Short Term Memory (LSTM) is introduced. 

LSTM networks are a special kind of RNN, capable of learning long-term dependencies. LSTMs are explicitly designed to avoid the long-term dependency problem. Remembering the information for long periods of time is their default behavior. So we used LSTM model for machine translation task, which involves huge long term memory and dependencies. 


From this-"The use of machine translation algorithm based on residual and LSTM neural network in translation teaching"
research article we used our baseline for the machine translation idea of using LSTM model. In this article they have used SCN-LSTM model for language translation.

On top of the paper we have mentioned above, after doing more research, we figured that LSTM alone is not good enough for language translation. So we ended up using a model based on paper ‘Sequence to Sequence learning with neural networks’ which is built on top of LSTM neural networks. 

We tried different approaches but a simple sequence to sequence model with the help of keras is the implementation that we ended up using for the project. Even with this simple implementation it gave pretty decent results. 

When we use just the LSTM to map a sequence of words from one language to another, it needs to have the same length of words for input and the output, but this is a rare case for any translation. Also the reference window LSTM model might be larger than RNN but it is not good enough to refer to sequences before that window. So encoder-decoder architecture, also called sequence to sequence learning with neural networks does a better job than single LSTM.  Encoder maps an input sequence into an abstract continuous representation that holds all the learned information of that given input. Decoder then takes the continuous representation and then step by step generates single output while also being fed to previous output. The encoder and decoder are usually layers of LSTMs or GRU. In our case we used LSTM. 


A seq2seq model requires to convert both the input and the output into integers of fixed length. For that purpose we vectorized our text data using keras’s Tokenizer() class. It turned sentences into sequences of integers, which were then padded with zeros to make all the sequences of the same length. So for the modeling part ,we used an embedding layer and an LSTM layer and for the decoder we used an LSTM layer followed by a dense layer. 


```python
# Seq2seq model
from keras.models import Sequential
from keras.layers import Embedding, Dense, LSTM, RepeatVector, TimeDistributed
from keras import optimizers
from keras.callbacks import EarlyStopping

def build_model(src_vocab, tar_vocab, src_timesteps, tar_timesteps, n_units):
    model = Sequential()
    model.add(Embedding(src_vocab, n_units, input_length=src_timesteps, mask_zero=True))
    model.add(LSTM(n_units))
    model.add(RepeatVector(tar_timesteps))
    model.add(LSTM(n_units, return_sequences=True))
    model.add(TimeDistributed(Dense(tar_vocab, activation='softmax')))
    return model
```

```python
# SGD Optimizer
model = build_model(src_vocab_size, tar_vocab_size, src_length, tar_length, 256)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# sgd = optimizers.SGD(lr=0.01)  
# model.compile(optimizer=sgd, loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model, to_file='model.png', show_shapes=True)
```

The above flow chart represents the encoder-decoder architecture, also called sequence to sequence learning with neural networks. The encoder and decoder are layers of LSTMs. For the modeling part, we used an embedding layer and an LSTM layer and for the decoder we used an LSTM layer followed by a dense layer. 

```python
# Fit the model

history = model.fit(trainX, trainY, epochs=20, batch_size=64, validation_split=0.2, verbose=1,
          callbacks=[EarlyStopping(
                    monitor='val_loss',
                    patience=10,
                    restore_best_weights=True
                    )])
```

```python
# Train & Test accuracy

tra, train_acc = model.evaluate(trainX, trainY, verbose=0)
tes, test_acc = model.evaluate(testX, testY, verbose=0)
print("Training Accuracy: ", train_acc)
print("Testing Accuracy: ", test_acc)
```

```python
def word_for_id(integer, tokenizer):
    # map an integer to a word
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None
```

```python
def predict_seq(model, tokenizer, source):
    # generate target from a source sequence
    prediction = model.predict(source, verbose=0)[0]
    integers = [np.argmax(vector) for vector in prediction]
    target = list()
    for i in integers:
        word = word_for_id(i, tokenizer)
        if word is None:
            break
        target.append(word)
    return ' '.join(target)
```

```python
# Model prediction

def compare_prediction(model, tokenizer, sources, raw_dataset, limit=10):
    actual, predicted = [], []
    src = f'{source_str.upper()} (SOURCE)'
    tgt = f'{target_str.upper()} (TARGET)'
    pred = f'LANGUAGE TRANSLATION IN {target_str.upper()}'
    print(f'{src:30} {tgt:25} {pred}\n')
    
    for i, source in enumerate(sources): # translate encoded source text
        source = source.reshape((1, source.shape[0]))
        translation = predict_seq(model, tar_tokenizer, source)
        raw_target, raw_src = raw_dataset[i]
        print(f'{raw_src:30} {raw_target:25} {translation}')
        if i >= limit: # Display some of the result
            break
```

# What is BLEU Score?

It can be used to evaluate your generated text against a reference text. NLTK provides the sentence_bleu() function for evaluating a generated sentence against one or more reference sentences. BLEU computes the modified precision metric using n-grams. But one problem with BLEU scores is that they tend to favor short translations but not long.

```python
# BLEU Score
from nltk.translate.bleu_score import corpus_bleu

def bleu_score(model, tokenizer, sources, raw_dataset):
    # Get the bleu score of a model
    actual, predicted = [], []
    for i, source in enumerate(sources):
        # translate encoded source text
        source = source.reshape((1, source.shape[0]))
        translation = predict_seq(model, tar_tokenizer, source)
        raw_target, raw_src = raw_dataset[i]
        actual.append([raw_target.split()])
        predicted.append(translation.split())
        
    bleu_dic = {}
    bleu_dic['1-grams'] = corpus_bleu(actual, predicted, weights=(1.0, 0, 0, 0))
    bleu_dic['1-2-grams'] = corpus_bleu(actual, predicted, weights=(0.5, 0.5, 0, 0))
    bleu_dic['1-3-grams'] = corpus_bleu(actual, predicted, weights=(0.3, 0.3, 0.3, 0))
    bleu_dic['1-4-grams'] = corpus_bleu(actual, predicted, weights=(0.25, 0.25, 0.25, 0.25))
    
    return bleu_dic
```

```python
# Compute the BLEU Score
bleu_train = bleu_score(model, tar_tokenizer, trainX, train)
bleu_test = bleu_score(model, tar_tokenizer, testX, test)
```

```python
plt.bar(x = bleu_train.keys(), height = bleu_train.values())
plt.title("BLEU Score with the training set")
plt.ylim((0,1))
plt.show()
```

High BLEU score is better. We can see that our model performed better in 1-gram, 2-gram & 3-gram comparisons with atmost 0.8 score and to the least of 0.3 in 4-gram translation, the score was lower when compared to others. The high scores denote the model’s better translation.

```python
plt.bar(x = bleu_test.keys(), height = bleu_test.values())
plt.title("BLEU Score with the test set")
plt.ylim((0,1))
plt.show()
```

We can see that our model performed better in 1-gram, 2-gram & 3-gram comparisons with atmost 0.7 score and to the least of 0.2 in 4-gram translation, the score was lower when compared to others. The high scores denote the model’s better translation. The model's score decreased when we added more words to compare during the translation task. May be it affected the co-occurence of the words.

# What is ROUGE score ?

ROUGE stands for Recall-Oriented Understudy for Gisting Evaluation. It is a set of metrics for evaluating machine translation. It works by comparing an automatically produced translation against a set of reference text/sentence (typically human-produced).


ROUGE-1 refers to the overlap of unigrams between the predicted result and the actual result.

In short, 

ROUGE-n recall=40% means that 40% of the n-grams in the reference summary are also present in the generated summary.
ROUGE-n precision=40% means that 40% of the n-grams in the generated summary are also present in the reference summary.

```python
!pip install rouge-score
```

```python
# ROUGE Score
from rouge_score import rouge_scorer

def rouge_score(model, tokenizer, sources, raw_dataset):
    # Get the ROUGE score of a model
    actual, predicted = [], []
    for i, source in enumerate(sources):
        # translate encoded source text
        source = source.reshape((1, source.shape[0]))
        translation = predict_seq(model, tar_tokenizer, source)
        raw_target, raw_src = raw_dataset[i]
        actual.append([raw_target.split()])
        predicted.append(translation.split())
        
#     rouge_dic = {}
    actual = ''.join(str(e) for e in actual)
    predicted = ''.join(str(e) for e in predicted)
    scorer = rouge_scorer.RougeScorer(['rouge1'], use_stemmer=True)
    scores = scorer.score(actual, predicted)
    
    return scores
```

```python
# Compute the ROUGE Score on train set 
rouge_train = rouge_score(model, tar_tokenizer, trainX, train)
rouge_train
```

```python
# Compute the ROUGE Score on test set 
rouge_test = rouge_score(model, tar_tokenizer, testX, test)
rouge_test
```

We used ROUGE-1 which refers to the overlap of unigrams between the predicted result and the actual result. If we see the precision & recall scores in the test set, the model crossed and translated around 85% of data correctly which is in the reference & generated texts / sentences. Also the F1-score shows good model performance.

```python
# Prepare training data
trainX = encode_sequences(src_tokenizer, src_length, train[:, idx_src])
trainY = encode_sequences(tar_tokenizer, tar_length, train[:, idx_tar])
trainY = encode_output(trainY, tar_vocab_size)

# Prepare test data
testX = encode_sequences(src_tokenizer, src_length, test[:, idx_src])
testY = encode_sequences(tar_tokenizer, tar_length, test[:, idx_tar])
testY = encode_output(testY, tar_vocab_size)
```

```python
# Seq2seq model
from keras.models import Sequential
from keras.layers import Embedding, Dense, LSTM, RepeatVector, TimeDistributed
from keras import optimizers
from keras.callbacks import EarlyStopping

def build_model(src_vocab, tar_vocab, src_timesteps, tar_timesteps, n_units):
    model = Sequential()
    model.add(Embedding(src_vocab, n_units, input_length=src_timesteps, mask_zero=True))
    model.add(LSTM(n_units))
    model.add(RepeatVector(tar_timesteps))
    model.add(LSTM(n_units, return_sequences=True))
    model.add(TimeDistributed(Dense(tar_vocab, activation='softmax')))
    return model
```

```python
# RMS Prop Optimizer
model = build_model(src_vocab_size, tar_vocab_size, src_length, tar_length, 256)
rms = optimizers.RMSprop(lr=0.001)  
# model.compile(optimizer=rms, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.compile(optimizer=rms, loss='categorical_crossentropy', metrics=['accuracy'])

print(model.summary())
```

The above flow chart represents the encoder-decoder architecture, also called sequence to sequence learning with neural networks. The encoder and decoder are layers of LSTMs. For the modeling part, we used an embedding layer and an LSTM layer and for the decoder we used an LSTM layer followed by a dense layer.

```python
def word_for_id(integer, tokenizer):
    # map an integer to a word
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None
```

```python
def predict_seq(model, tokenizer, source):
    # generate target from a source sequence
    prediction = model.predict(source, verbose=0)[0]
    integers = [np.argmax(vector) for vector in prediction]
    target = list()
    for i in integers:
        word = word_for_id(i, tokenizer)
        if word is None:
            break
        target.append(word)
    return ' '.join(target)
```

# What is BLEU Score?

It can be used to evaluate your generated text against a reference text. NLTK provides the sentence_bleu() function for evaluating a generated sentence against one or more reference sentences. BLEU computes the modified precision metric using n-grams. But one problem with BLEU scores is that they tend to favor short translations but not long.

```python
# BLEU Score

def bleu_score(model, tokenizer, sources, raw_dataset):
    # Get the bleu score of a model
    actual, predicted = [], []
    for i, source in enumerate(sources):
        # translate encoded source text
        source = source.reshape((1, source.shape[0]))
        translation = predict_seq(model, tar_tokenizer, source)
        raw_target, raw_src = raw_dataset[i]
        actual.append([raw_target.split()])
        predicted.append(translation.split())
        
    bleu_dic = {}
    bleu_dic['1-grams'] = corpus_bleu(actual, predicted, weights=(1.0, 0, 0, 0))
    bleu_dic['1-2-grams'] = corpus_bleu(actual, predicted, weights=(0.5, 0.5, 0, 0))
    bleu_dic['1-3-grams'] = corpus_bleu(actual, predicted, weights=(0.3, 0.3, 0.3, 0))
    bleu_dic['1-4-grams'] = corpus_bleu(actual, predicted, weights=(0.25, 0.25, 0.25, 0.25))
    
    return bleu_dic
```

```python
# Compute the BLEU Score
bleu_train = bleu_score(model, tar_tokenizer, trainX, train)
bleu_test = bleu_score(model, tar_tokenizer, testX, test)
```

```python
# ROUGE score
from rouge_score import rouge_scorer

def rouge_score(model, tokenizer, sources, raw_dataset):
    # Get the ROUGE score of the model
    actual, predicted = [], []
    for i, source in enumerate(sources):
        # translate encoded source text
        source = source.reshape((1, source.shape[0]))
        translation = predict_seq(model, tar_tokenizer, source)
        raw_target, raw_src = raw_dataset[i]
        actual.append([raw_target.split()])
        predicted.append(translation.split())
        
#     rouge_dic = {}
    actual = ''.join(str(e) for e in actual)
    predicted = ''.join(str(e) for e in predicted)
    scorer = rouge_scorer.RougeScorer(['rouge1'], use_stemmer=True)
    scores = scorer.score(actual, predicted)
    
    return scores
```

# What is ROUGE score ?

ROUGE stands for Recall-Oriented Understudy for Gisting Evaluation. It is a set of metrics for evaluating machine translation. It works by comparing an automatically produced translation against a set of reference text/sentence (typically human-produced).


ROUGE-1 refers to the overlap of unigrams between the predicted result and the actual result.

In short, 

ROUGE-n recall=40% means that 40% of the n-grams in the reference summary are also present in the generated summary.
ROUGE-n precision=40% means that 40% of the n-grams in the generated summary are also present in the reference summary.
